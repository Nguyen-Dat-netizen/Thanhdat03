```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.common.exceptions import NoSuchElementException, TimeoutException
import time

# Configure Chrome Options
chrome_options = webdriver.ChromeOptions()
# chrome_options.add_argument('--headless')  # Uncomment to run in headless mode
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage')
chrome_options.add_argument("user-agent=Your User Agent String Here")
chrome_options.add_argument('--disable-blink-features=AutomationControlled')

# Base URL
base_url = "https://cellphones.com.vn/laptop.html"

# Initialize driver
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=chrome_options)

try:
    # Open the webpage
    driver.get(base_url)

    # Click "Xem thêm" buttons (with error handling)
    while True:
        time.sleep(2)  # Wait for the page to load
        try:
            view_more_buttons = driver.find_elements(By.XPATH, "/html/body/div[1]/div/div/div[3]/div[2]/div/div[8]/div[5]/div/div[2]/a")

            # Break if no more buttons found
            if not view_more_buttons:
                print("No more 'Xem thêm' buttons to click.")
                break

            # Click each button
            for button in view_more_buttons:
                button.click()
                time.sleep(2)  # Allow time for products to load

                # Try to close ad if it appears
                time.sleep(1)  # Wait a moment before checking for ad
                try:
                    ad_close_button = driver.find_element(By.CSS_SELECTOR, ".cancel-button-top")
                    ad_close_button.click()
                    time.sleep(1)  # Wait after closing the ad
                except NoSuchElementException:
                    print("No ad close button found.")

        except NoSuchElementException as e:
            print(f"Error clicking 'Xem thêm' buttons: {e}")
            break

    # Get product names
    time.sleep(2)  # Wait for products to load
    product_elements = driver.find_elements(By.CSS_SELECTOR, ".product__name")
    product_names = [product.text for product in product_elements]
    product_names = product_names[10:]
    
    product_links = driver.find_elements(By.CSS_SELECTOR, ".product-info [href]")
    links = [product.get_attribute('href') for product in product_links]
    links = links[10:]
    
    product_prices= driver.find_elements(By.CSS_SELECTOR, ".product__price--show")
    product_price = [product.text for product in product_prices]
    product_price = product_price[10:]
    
    df1=pd.DataFrame(list(zip(product_names,links,product_price)),columns=['product_names','links','product_price'])
    
    df1['index_']=np.arange(1,len(df1)+1)
    
finally:
    # Close the driver
    #driver.quit()
    pass
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.common.exceptions import NoSuchElementException, TimeoutException
import time

# Configure Chrome Options
chrome_options = webdriver.ChromeOptions()
# chrome_options.add_argument('--headless')  # Uncomment to run in headless mode
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage')
chrome_options.add_argument("user-agent=Your User Agent String Here")
chrome_options.add_argument('--disable-blink-features=AutomationControlled')

# Base URL
base_url = "https://cellphones.com.vn/laptop.html"

# Initialize driver
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=chrome_options)

try:
    # Open the webpage
    driver.get(base_url)

    # Click "Xem thêm" buttons (with error handling)
    while True:
        time.sleep(2)  # Wait for the page to load
        try:
            view_more_buttons = driver.find_elements(By.XPATH, "/html/body/div[1]/div/div/div[3]/div[2]/div/div[8]/div[5]/div/div[2]/a")

            # Break if no more buttons found
            if not view_more_buttons:
                print("No more 'Xem thêm' buttons to click.")
                break

            # Click each button
            for button in view_more_buttons:
                button.click()
                time.sleep(2)  # Allow time for products to load

                # Try to close ad if it appears
                time.sleep(1)  # Wait a moment before checking for ad
                try:
                    ad_close_button = driver.find_element(By.CSS_SELECTOR, ".cancel-button-top")
                    ad_close_button.click()
                    time.sleep(1)  # Wait after closing the ad
                except NoSuchElementException:
                    print("No ad close button found.")

        except NoSuchElementException as e:
            print(f"Error clicking 'Xem thêm' buttons: {e}")
            break

    # Get product names
    time.sleep(2)  # Wait for products to load
    
    price_old_list, price_old_dix = [], []

    for i in range(1, len(product_names) + 1):
        try:
            # Adjust the XPath to match the structure of your HTML
            price = driver.find_element(By.XPATH, f"//div[{i}]/div[1]/a/div[3]/div[2]/div/p")
            price_old_list.append(price.text)
            print(i)
            price_old_dix.append(i)
        except NoSuchElementException:
            print(f"Null {i}")
            
    discount_list, discount_dix = [], []

    for i in range(1, len(product_names) + 1):
        try:
            # Adjust the XPath to match the structure of your HTML
            discount = driver.find_element(By.XPATH, f"//div[{i}]/div[1]/a/div[3]/div[2]/div/div/p")
            discount_list.append(discount.text)
            print(i)
            discount_dix.append(i)
        except NoSuchElementException:
            print(f"Null {i}") 
            
    discount_list_Student, discount_dix_Student = [], []

    for i in range(1, len(product_names) + 1):
        try:
            # Adjust the XPath to match the structure of your HTML
            discount = driver.find_element(By.XPATH, f"//div[{i}]/div[1]/a/div[5]//span")
            discount_list_Student.append(discount.text)
            print(i)
            discount_dix_Student.append(i)
        except NoSuchElementException:
            print(f"Null {i}") 
            
    
    
finally:
    # Close the driver
    #driver.quit()
    pass
```

```python
df2=pd.DataFrame(list(zip(price_old_dix,price_old_list)),columns=['price_old_dix','price_old_list'])
df3=pd.DataFrame(list(zip(discount_dix,discount_list)),columns=['discount_dix','discount_list'])
df4=pd.DataFrame(list(zip(discount_dix_Student,discount_list_Student)),columns=['discount_dix_Student','discount_list_Student'])
```

```python
df5=df1.merge(df2,how='left',left_on='index_',right_on='price_old_dix')
df6=df5.merge(df3,how='left',left_on='index_',right_on='discount_dix')
df7=df6.merge(df4,how='left',left_on='index_',right_on='discount_dix_Student')
```
