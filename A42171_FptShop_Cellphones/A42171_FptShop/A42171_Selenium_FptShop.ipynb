```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.common.exceptions import NoSuchElementException, TimeoutException
import time

# Configure Chrome Options
chrome_options = webdriver.ChromeOptions()
# chrome_options.add_argument('--headless')  # Uncomment to run in headless mode
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage')
chrome_options.add_argument("user-agent=Your User Agent String Here")
chrome_options.add_argument('--disable-blink-features=AutomationControlled')

# Base URL
base_url = "https://fptshop.com.vn/may-tinh-xach-tay"

# Initialize driver
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=chrome_options)

try:
    # Open the webpage
    driver.get(base_url)

    # Click "Xem thêm" buttons (with error handling)
    while True:
        time.sleep(2)  # Wait for the page to load
        try:
            view_more_buttons = driver.find_elements(By.XPATH, "/html/body/main/div/div[6]/div[2]/div[3]/div[3]/button")

            # Break if no more buttons found
            if not view_more_buttons:
                print("No more 'Xem thêm' buttons to click.")
                break

            # Click each button
            for button in view_more_buttons:
                button.click()
                time.sleep(2)  # Allow time for products to load

                # Try to close ad if it appears
                time.sleep(1)  # Wait a moment before checking for ad
        except NoSuchElementException as e:
            print(f"Error clicking 'Xem thêm' buttons: {e}")
            break

    # Get product names
    time.sleep(2)  # Wait for products to load
    price_old_list, price_old_dix = [], []

    for i in range(1, len(product_names) + 1):
        try:
            # Adjust the XPath to match the structure of your HTML
            price = driver.find_element(By.XPATH, f"//div[{i}]/div/div[2]/div[1]/p[1]/span[1]")
            price_old_list.append(price.text)
            print(i)
            price_old_dix.append(i)
        except NoSuchElementException:
            print(f"Null {i}")
    
    discount_list, discount_dix = [], []

    for i in range(1, len(product_names) + 1):
        try:
            # Adjust the XPath to match the structure of your HTML
            discount = driver.find_element(By.XPATH, f"//div[{i}]/div/div[2]/div[1]/p[1]/span[2]")
            discount_list.append(discount.text)
            print(i)
            discount_dix.append(i)
        except NoSuchElementException:
            print(f"Null {i}") 
    
    
    discount_list1, discount_dix1 = [], []

    for i in range(1, len(product_names) + 1):
        try:
            # Adjust the XPath to match the structure of your HTML
            discount = driver.find_element(By.XPATH, f"//div[{i}]/div/div[2]/div[1]/p[3]")
            discount_list1.append(discount.text)
            print(i)
            discount_dix1.append(i)
        except NoSuchElementException:
            print(f"Null {i}") 
    
    
finally:
    # Close the driver
    #driver.quit()
    pass
```

```python
df2=pd.DataFrame(list(zip(price_old_dix,price_old_list)),columns=['price_old_dix','price_old_list'])
df3=pd.DataFrame(list(zip(discount_dix,discount_list)),columns=['discount_dix','discount_list'])
df4=pd.DataFrame(list(zip(discount_dix1,discount_list1)),columns=['discount_dix1','discount_list1'])
```

```python
df5=df1.merge(df2,how='left',left_on='index_',right_on='price_old_dix')
df6=df5.merge(df3,how='left',left_on='index_',right_on='discount_dix')
df7=df6.merge(df4,how='left',left_on='index_',right_on='discount_dix1')
```
