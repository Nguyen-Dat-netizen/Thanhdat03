```python
#Lấy ra số Qùa cho mỗi sản phẩm được thưởng
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException

Gift_dix, Gift_list = [], []

for i in range(1, len(data_name) + 1):
    try:
        # Adjust the XPath to match the structure of your HTML
        Gift = driver.find_element(By.XPATH, f"//li[{i}]/a/p[2]/b")
        Gift_list.append(Gift.text)
        print(i)
        Gift_dix.append(i)
    except NoSuchElementException:
        print(f"Null {i}") 
```

```python
#hoan thien lấy ra Qùa
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException

Gift_dix, Gift_list = [], []
xpaths = [
    lambda i: f"//li[{i}]/a/p[2]/b", 
    lambda i: f"//li[{i}]/a/p/b" 
]

for i in range(1, len(data_name) + 1):
    found = False
    for xpath in xpaths:
        try:
            Gift = driver.find_element(By.XPATH, xpath(i)) 
            Gift_list.append(Gift.text)
            Gift_dix.append(i)
            print(f"Quà tặng ở vị trí {i}: {Gift.text}")
            found = True
            break  # Thoát khỏi vòng lặp nếu tìm thấy
        except NoSuchElementException:
            continue  # Tiếp tục thử XPath tiếp theo

    if not found:
        print(f"Không tìm thấy quà tặng ở vị trí {i}")
```

```python
df6=pd.DataFrame(list(zip(Gift_dix,Gift_list)),columns=['Gift_dix','Gift_list'])
```

```python
df7=df5.merge(df6,how='left',left_on='index_',right_on='Gift_dix')
```
