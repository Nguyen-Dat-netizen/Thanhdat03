# Chuyển dữ liệu cột "price" sang kiểu float

```python
data['price'] = data['price'].str.replace('.', '', regex=False).str.replace('₫', '', regex=False).astype(float)
```

# Chuyển dữ liệu cột "discount_list" sang kiểu float

```python
data['discount_list'] = data['discount_list'].str.replace('.', '', regex=False).str.replace('₫', '', regex=False).astype(float)
```

# Chuyển dữ liệu cột "Gift_list" sang kiểu float

```python
data['Gift_list'] = data['Gift_list'].str.replace('.', '', regex=False).str.replace('₫', '', regex=False).fillna(0).astype(float)
```

# Chuyển những giá trị "Evaluate_list" NaN thay bằng 0

```python
data['Evaluate_list'] = pd.to_numeric(data['Evaluate_list'], errors='coerce').fillna(0).astype(int)
```

# Tính lại cột "discount_percent"
    Nhận xét: Những giá trị "discount_percent" bị sai và Những giá trị còn bị trống
    Giai Pháp: Tính lại bằng công thức : data['discount_percent'].fillna(
    (data['discount_list'] - data['price']) / data['discount_list']

```python
data['discount_percent'] = pd.to_numeric(data['discount_percent'], errors='coerce')
data['discount_percent'] = data['discount_percent'].fillna(
    (data['discount_list'] - data['price']) / data['discount_list']
)*100 # Nhân với 100 để chuyển sang phần trăm
```

#  Chuyển những giá trị Evaluate_count_list về kiểu int và những giá trị NaN sẽ chuyển thành 0 

```python
data['Evaluate_count_list'] = (
    data['Evaluate_count_list']
    .str.replace(' đánh giá', '', regex=False)  # Xóa phần văn bản ' đánh giá'
    .str.replace('Đánh giá', '', regex=False)  # Xóa phần văn bản 'Đánh giá'
)

# Chuyển đổi thành kiểu số nguyên, với xử lý trường hợp lỗi
data['Evaluate_count_list'] = pd.to_numeric(data['Evaluate_count_list'], errors='coerce').fillna(0).astype(int)
```

## Giai pháp: đối với những giá trị "discount_list" mà bằng NaN thì ta cho những giá trị đấy bằng "price" và Khi đó "discount_percent" sẽ bằng 0

```python
data.loc[data['discount_list'].isna() & data['discount_percent'].isna(), 'discount_list'] = data['price']
data.loc[data['discount_percent'].isna(), 'discount_percent'] = 0
```
